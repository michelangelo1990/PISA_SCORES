## Osvrt na PISA score

Source: Organization for Economic Cooperation and Development (OECD), 2015-2016

The Program for International Student Assessment (PISA) is a worldwide study by OECD in 70 nations of 15-year-old  studentsâ€™ scholastic performance on mathematics, science and reading.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data = pd.read_excel("PISA_SCORE.xlsx",header = 0,dtype={'NUM': str,'COUNTRY':str,'AVG': float})
```

```python
data.head()
```

```python
data.tail()
```

```python
dataframe = data[['COUNTRY','AVG']]
```

```python
dataframe.head()
```

```python
plt.subplots(figsize=(20,15))
ax = sns.barplot(x="AVG", y="COUNTRY", data=dataframe)
locs, labels = plt.yticks()
labels[35]=r'$\mathbf{CROATIA}$'
plt.yticks(locs, labels)
plt.title('Odnos prosjeÄnih PISA rezultata (ukupno) prema drÅ¾avama')
plt.xlabel('ProsjeÄne vrijednosti')
plt.ylabel('DrÅ¾ava')
plt.show()
```

```python
cro_score = dataframe[dataframe.COUNTRY=='Croatia']
cro_score
```

```python
from matplotlib.lines import Line2D
plt.subplots(figsize=(8,6))
plt.title('Distribucija PISA score-ova')
ax = sns.distplot(dataframe.AVG, bins=15)
ax.axvline(cro_score.AVG[35],color='red')
plt.xlabel('ProsjeÄna vrijednost')
ticks = ax.xaxis.get_majorticklocs()
ax.xaxis.set_ticks(np.append(ticks,cro_score.AVG[35]))
plt.xticks(rotation=60)
plt.legend(handles=[Line2D([0], [0], color='red', lw=2, label='Hrvatska')])
plt.show()
```

```python
from matplotlib.lines import Line2D
plt.subplots(figsize=(8,6))
plt.title('Distribucija PISA score-ova')
ax = sns.distplot(dataframe.AVG, bins=15)
ax.axvline(cro_score.AVG[35],color='red')
ax.axvline(dataframe.AVG.mean(),color='green')
ax.axvline(dataframe.AVG.median(),color='yellow')
plt.xlabel('ProsjeÄna vrijednost')
ticks = ax.xaxis.get_majorticklocs()
ax.xaxis.set_ticks(np.append(ticks,cro_score.AVG[35]))
plt.xticks(rotation=60)
plt.legend(handles=[Line2D([0], [0], color='red', lw=2, label='Hrvatska'),Line2D([0], [0], color='green', lw=2, label='Prosjek'),Line2D([0], [0], color='yellow', lw=2, label='Medijan')])
plt.show()
```

```python
data_math = pd.read_excel("PISA_MATH.xlsx",header = 0,dtype={'COUNTRY':str,'AVG_MATH': float})
data_math.head()
```

```python
plt.subplots(figsize=(20,15))
ax = sns.barplot(x="AVG_MATH", y="COUNTRY", data=data_math)
locs, labels = plt.yticks()
labels[40]=r'$\mathbf{CROATIA}$'
plt.yticks(locs, labels)
plt.title('Odnos prosjeÄnih PISA rezultata (matematika) prema drÅ¾avama')
plt.xlabel('ProsjeÄne vrijednosti')
plt.ylabel('DrÅ¾ava')
plt.show()
```

```python
cro_score_math = data_math[data_math.COUNTRY=='Croatia']
cro_score_math
```

```python
from matplotlib.lines import Line2D
plt.subplots(figsize=(8,6))
plt.title('Distribucija PISA score-ova u matematici')
ax = sns.distplot(data_math.AVG_MATH, bins=15)
ax.axvline(cro_score_math.AVG_MATH[40],color='red')
ax.axvline(data_math.AVG_MATH.mean(),color='green')
ax.axvline(data_math.AVG_MATH.median(),color='yellow')
plt.xlabel('ProsjeÄna vrijednost')
ticks = ax.xaxis.get_majorticklocs()
ax.xaxis.set_ticks(np.append(ticks,cro_score_math.AVG_MATH[40]))
plt.xticks(rotation=60)
plt.legend(handles=[Line2D([0], [0], color='red', lw=2, label='Hrvatska'),Line2D([0], [0], color='green', lw=2, label='Prosjek'),Line2D([0], [0], color='yellow', lw=2, label='Medijan')])
plt.show()
```
